```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.bbc.com/"  # Replace with your target URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    print("Successfully fetched the website!")
else:
    print("Failed to fetch the website")
```

```python
# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")

# Print the parsed HTML (Optional: useful for inspecting the structure)
print(soup.prettify())
```

```python
# Find all <a> tags and extract the href attribute (links)
links = soup.find_all("a")

for link in links:
    href = link.get("href")
    if href:
        print(href)
```

```python
title = soup.title.text
print("Page Title:", title)

# Extract all headings (h1 tags)
headings = soup.find_all("h1")
for h1 in headings:
    print("Heading:", h1.text)

# Extract all paragraphs (p tags)
paragraphs = soup.find_all("p")
for p in paragraphs:
    print("Paragraph:", p.text)
```
